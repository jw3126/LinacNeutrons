```julia
using Revise
using Plots
using PhaseSpaceIO

path = joinpath("output", "phspLinac.IAEAheader")
particles = open_phsp(collect, path);
@show length(particles)
```

```julia
neutrons = filter(isneutron, particles)
photons = filter(isphoton, particles)
electrons = filter(iselectron, particles)
positrons = filter(ispositron, particles)
energy(p) = p.E
@show length(electrons)
# @show sum(energy, electrons)
@show length(positrons)
# @show sum(energy, positrons)
@show length(photons)
# @show sum(energy, photons)
@show length(neutrons)
# @show sum(energy, neutrons)
histogram(map(p->p.E, neutrons))
```

```julia
histogram(map(energy, neutrons))
```
